```python
import requests
from json import JSONDecodeError
import pickle
```

```python
def get_json_entity_by_id(id_val):
    int_string = id_val
    base_url = 'https://cosylab.iiitd.edu.in/flavordb/entities_json?id='
    url = f'{base_url}{int_string}'
    result = requests.get(url)
    try:
        info = result.json()
    except JSONDecodeError:
        info = {}
    return info
```

```python
flavordb = {}
for i in range(2000):
    flavordb[i] = get_json_entity_by_id(i)
```

```python
with open('../data/external/flavor_db.pkl', 'wb') as f:
    pickle.dump(flavordb, f)
```

```python
import pandas as pd
```

```python
flavordb_info = pd.DataFrame(flavordb).T
flavor_db_df = flavordb_info[flavordb_info.isna().sum(axis=1) < 11].reset_index(drop=True)
```

```python
flavor_db_df.to_csv('../data/external/flavor_db.csv')
```
