# How is the information related?
In this notebook we made an analysis, by hand on the relation between the USDA datasets. 

## Imports

```python
import networkx as nx
import pandas as pd
```

## Create graph

```python
usda_data = nx.Graph()

food__edge_list = [
    ('food', 'branded_food'), 
    ('food', 'sub_sample_food'),
    ('food', 'sr_legacy_food'), 
    ('food', 'market_acquistion'),
    ('food', 'survey_fndds_food'),
    ('food', 'sample_food'),
    ('food', 'agricultural_acquisition'),
    ('food', 'foundation_food'),
    ('food', 'experimental_food'),    
]

acquisition_samples__edge_list = [
    ('sample_food', 'market_acquistion'),
    ('input_food', 'survey_fndds_food'),
    ('food_nutrient', 'branded_food'),
    ('food_nutrient', 'sub_sample_food'),
    ('food_nutrient', 'sr_legacy_food'),
    ('food_nutrient', 'survey_fndds_food'),
    ('food_nutrient', 'agricultural_acquisition'),
    ('food_nutrient', 'foundation_food'),
    ('food_nutrient', 'food_nutrient_derivation'),
    ('food_nutrient', 'nutrient'),
    ('food_nutrient', 'nutrient_incoming_name'),
    ('nutrient_incoming_name', 'nutrient'),



    ('food_portion', 'sr_legacy_food'),
    ('food_portion', 'survey_fndds_food'),
    ('food_portion', 'market_acquistion'),
    ('food_portion', 'foundation_food'),
    
    ('food_component', 'market_acquistion'),
    ('food_component', 'foundation_food'),
    ('food_attribute', 'branded_food'),
    ('food_attribute', 'survey_fndds_food'),
    ('food_attribute', 'sr_legacy_food'),
    ('food_attribute', 'agricultural_acquisition'),
    ('food_attribute', 'market_acquistion'),
    ('food_attribute', 'foundation_food'),
]

usda_data.add_edges_from(food__edge_list)
usda_data.add_edges_from(acquisition_samples__edge_list)
pos = nx.spring_layout(usda_data)
nx.draw(usda_data,pos=pos, with_labels=True)
```
