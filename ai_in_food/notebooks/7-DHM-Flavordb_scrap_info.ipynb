# Obtener la información de FlavorDb:

¿acaso podemos obtener la informacion de FlavorDb? Posiblemente se pueda a través de WebScrapping. 

## Imports

```python
import pandas as pd
import pickle
import requests

from json import JSONDecodeError
```

## Functions

```python
def get_json_entity_by_id(id_val):
    int_string = id_val
    base_url = 'https://cosylab.iiitd.edu.in/flavordb/entities_json?id='
    url = f'{base_url}{int_string}'
    result = requests.get(url)
    try:
        info = result.json()
    except JSONDecodeError:
        info = {"error": id_val}
    return info
```

## Scrap
Vamos a obtener el json correspondiente a traves de web_scrapping, obteniendo el valor del índice `id_val`.

Para esto se utilizará el siguiente link: https://cosylab.iiitd.edu.in/flavordb/entities_json?id=`id_val`

```python
flavordb = {} # Aquí se guarda la información del json. 
search_vals_max = 974 # El valor de 974 se incluye por si hubiese un valor que cupiera
for index_val in range(search_vals_max): 
    flavordb[index_val] = get_json_entity_by_id(index_val)
```

```python
# Filter the case when error is in a key. 
for key in list(flavordb.keys())[:]:
    information = flavordb[key]
    list_of_keys = list(information.keys())
    if 'error' in list_of_keys:
        del flavordb[key]
    
    
```

## Outputs:

```python
# Aquí guardamos flavor_db como un archivo pkl. 
with open('../data/external/flavor_db.pkl', 'wb') as f:
    pickle.dump(flavordb, f)
```

```python
# Revisamos que se haya guardado la info
with open('../data/external/flavor_db.pkl', 'rb') as f:
    f = pickle.load(f)
```
